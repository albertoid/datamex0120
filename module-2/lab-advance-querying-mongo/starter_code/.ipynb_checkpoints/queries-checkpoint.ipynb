```python
from pymongo import MongoClient
client=MongoClient()
db=client.companies
```

# Answers
### 1. All the companies that it's name match 'Babelgum'. Retrieve only their `name` field.

```python
name_Babelgum=db.companies.find({'name':'Babelgum'},
                                {'name':1,'_id':0})
list(name_Babelgum)
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
Egt5000=db.companies.find({'number_of_employees':{'$gt':5000}},
                           {'name':1,'_id':0}).limit(20)
list(Egt5000)
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fileds.

```python
Fb2000_2005=db.companies.find(
                        {'$and':[
                                  {'founded_year':{'$gte':2000}},
                                  {'founded_year':{'$lte':2005}}
                                ]
                        },
                        {'name':1,'_id':0,'founded_year':1}).limit(20)
list(Fb2000_2005)
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
VAgt100Mio=db.companies.find(
                        {'$and':[
                                  {'founded_year':{'$lte':2010}},
                                  {'ipo.valuation_amount':{'$gt':100000000}}
                                ]
                        },
                        {'name':1,'_id':0,'ipo':1})
list(VAgt100Mio)                          
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
El1kFb2005=db.companies.find(
                        {'$and':[
                                  {'number_of_employees':{'$lt':1000}},
                                  {'founded_year':{'$lt':2005}}
                                ]
                        },
                        {'name':1,'_id':0}).sort('founded_year',1).limit(10)
list(El1kFb2005)    
```

### 6. All the companies that don't include the `partners` field.

```python
El1kFb2005=db.companies.find({'partners':{'$exists':False}},
                        {'name':1,'_id':0})
list(El1kFb2005)
```

### 7. All the companies that have a null type of value on the `category_code` field.

```python
CCn=db.companies.find({'category_code':None},{'name':1,'_id':0})
list(CCn)
```

### 8. All the companies that have at least 100 employees but less than 1000. Retrieve only the `name` and `number of employees` fields.

```python
Egt100Elt1k=db.companies.find(
                        {'$and':[
                                  {'number_of_employees':{'$gt':100}},
                                  {'number_of_employees':{'$lt':1000}}
                                ]
                        },
                        {'name':1,'_id':0,'number_of_employees':1})
list(Egt100Elt1k)  
```

### 9. Order all the companies by their IPO price descendently.

```python
IPOd=db.companies.find({'ipo.valuation_amount':{'$exists':True}},
                      {'name':1,'_id':0} 
                      ).sort('ipo.valuation_amount',-1)
list(IPOd)
```

### 10. Retrieve the 10 companies with more employees, order by the `number of employees`

```python
ETop10=db.companies.find({},
                      {'name':1,'_id':0}#,'number_of_employees':1} 
                      ).sort('number_of_employees',-1).limit(10)
list(ETop10)
```

### 11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.

```python
FMss=db.companies.find(
                       {'founded_month':{'$gt':6}},
                       {'name':1,'_id':0}
                      ).limit(1000)
list(FMss)
```

### 12. All the companies that have been 'deadpooled' after the third year.

```python
dp3=db.companies.find({'$and':[
                              {'deadpooled_year':{'$exists':True},
                               'founded_year':{'$exists':True},
                               'deadpooled_year':{'$ne':None}}
                              ]},
                        {'name':1,'_id':0,'deadpooled_year':1,'founded_year':1} 
                        )
list(dp3)
```

### 13. All the companies founded before 2000 that have and acquisition amount of more than 10.000.000

```python
Fb2kAAm10m=db.companies.find({'$and':[
                              {'acquisition.price_amount':{'$gt':10000000},
                               'founded_year':{'$lt':2000},}
                              ]},
                        {'name':1,'_id':0} 
                        )
list(Fb2kAAm10m)
```

### 14. All the companies that have been acquired after 2015, order by the acquisition amount, and retrieve only their `name` and `acquisiton` field.

```python
 A2015=db.companies.find(
                        {'acquisition.acquired_year':{'$gt':2015}},
                        {'name':1,'_id':0,'acquisiton':1} 
                        ).sort('acquisition.price_amount',-1)
list(A2015)
```

### 15. Order the companies by their `founded year`, retrieving only their `name` and `founded year`.

```python
Onf=db.companies.find({'$and':
                      [
                      {'founded_year':{'$exists':True},
                       'founded_year':{'$ne':None}
                      }
                      ]},
                      {'name':1,'_id':0,'founded_year':1} 
                      ).sort('founded_year',1).limit(2000)
list(Onf)
```

### 16. All the companies that have been founded on the first seven days of the month, including the seventh. Sort them by their `aquisition price` descendently. Limit the search to 10 documents.

```python
F7d=db.companies.find({'$and':
                      [
                      {'founded_day':{'$let':7},
                       'founded_day':{'$ne':None}}
                      ]},
                      {'name':1,'_id':0} 
                      ).sort('aquisition_price',1).limit(10)
list(F7d)
```

### 17. All the companies on the 'web' `category` that have more than 4000 employees. Sort them by the amount of employees in ascendant order.

```python
F7d=db.companies.find({'$and':
                      [
                      {'category_code':'web',
                       'number_of_employees':{'$gt':4000}}
                      ]},
                      {'name':1,'_id':0} 
                      ).sort('number_of_employees',1)
list(F7d)
```

### 18. All the companies which their acquisition amount is more than 10.000.000, and currency are 'EUR'.


```python
AA10mCe=db.companies.find({'$and':
                      [
                      {'acquisition.price_amount':{'$gt':10000000},
                       'acquisition.price_currency_code':'EUR'}
                      ]},
                      {'name':1,'_id':0} 
                      )
list(AA10mCe)
```

### 19. All the companies that have been acquired on the first trimester of the year. Limit the search to 10 companies, and retrieve only their `name` and `acquisition` fields.

```python
A1t=db.companies.find(
                        {'$and':[
                                  {'acquisition.acquired_month':{'$exists':True}},
                                  {'acquisition.acquired_month':{'$lt':4}}
                                ]
                        },
                        {'name':1,'_id':0,'acquisition':1}).limit(10)
list(A1t)  
```

### 20. All the companies that have been founded between 2000 and 2010, but have not been acquired before 2011.

```python
F00_10A11=db.companies.find(
                        {'$and':[
                                  {'founded_year':{'$gte':2000}},
                                  {'founded_year':{'$lte':2010}},
                                  {'acquisition.acquired_year':{'$gte':2011}},
                                ]
                        },
                        {'name':1,'_id':0})
list(F00_10A11)  
```

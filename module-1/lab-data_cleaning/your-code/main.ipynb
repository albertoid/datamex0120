```python
#userId in posts and view posts to integers, so any of this values could be float number
#and all the numbers could be represented in 32 bits

mergedTable['userId']= mergedTable['userId'].astype('int16')
mergedTable['Reputation']= mergedTable['Reputation'].astype('int8')
mergedTable['Views']= mergedTable['Views'].astype('int32')
mergedTable['UpVotes']= mergedTable['UpVotes'].astype('int32')
mergedTable['DownVotes']= mergedTable['DownVotes'].astype('int32')
mergedTable['postId']= mergedTable['postId'].astype('int32')
mergedTable['Score']= mergedTable['Score'].astype('int32')
mergedTable['ViewCount']= mergedTable['ViewCount'].astype('int16')
mergedTable['CommentCount']= mergedTable['CommentCount'].astype('int16')

mergedTable.dtypes
```

#### Bonus: Identify extreme values in your merged dataframe as you have learned in class, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder.

```python
merged_df= mergedTable.copy()
```

```python
stats=merged_df.describe().T
stats['IQR']=stats['75%']-stats['25%']
stats
```

```python
outliers=pd.DataFrame(columns=merged_df.columns)
outliers

for e in stats.index:
    iqr=stats.at[e, 'IQR']
    cutoff=iqr*1.5
    lower=stats.at[e, '25%']-cutoff
    upper=stats.at[e, '75%']+cutoff
    results=merged_df[(merged_df[e]<lower)|(merged_df[e]>upper)].copy()
    results['outliers']=e
    outliers=outliers.append(results, sort=True)
    
outliers['outliers'].value_counts()
```

```python
clean_index=[i for i in merged_df.index if i not in outliers.index]
len(clean_index)
```

```python
df_superclean=pd.DataFrame(columns=merged_df.columns)
df_superclean=merged_df.iloc[clean_index, :]
len(df_superclean)
```

```python
#Carga CVS
dominio = 'https://coinmetrics.io/newdata/'
ext = ".csv"
crypto=['btc','mana','eth','gnt','ltc','xrp','tusd','bat']
data_csv=[]

for i in tqdm(range(len(crypto))):
    csv_response=requests.get(dominio+crypto[0]+ext)
    with open(crypto[i]+'.csv', 'w') as f:
        writer = csv.writer(f)
        for line in csv_response.iter_lines():
            writer.writerow(line.decode('utf-8').split(','))
    data_csv.append(pd.read_csv("./"+crypto[i]+".csv"))

```

```python
#Carga html
html = 'https://coinmarketcap.com/currencies/decentraland/historical-data/?start=20130428&end=20200126'
resp_html=requests.get(html).content
sopa=bs(resp_html,'lxml')
```

```python
def get_table_values(nombre,sopa,clase):
    nombre=sopa.find_all('td',{'class': clase})
    nombre= list(nombre)
    
    for i in range(len(nombre)):
            nombre[i]=nombre[i].find_all('div')
            nombre[i]=str(nombre[i])
            nombre[i]=nombre[i].replace('<div class="">','')
            nombre[i]=nombre[i].replace('','')
            nombre[i]=nombre[i].replace('</div>','')    
            
    return nombre
```

```python
fecha=[]
valores=[]
fecha = get_table_values(fecha,sopa,'cmc-table__cell cmc-table__cell--sticky cmc-table__cell--left')
valores = get_table_values(valores,sopa,'cmc-table__cell cmc-table__cell--right')
```

```python
meses={'Jan':'01','Feb':'02','Mar':'03','Apr':'04','May':'05','Jun':'06',
        'Jul':'07','Aug':'08','Sep':'09','Oct':'10','Nov':'11','Dec':'12'}
mes =''
dia = ''
año=''
for i in range(len(fecha)):
    mes = fecha[i][1:4]
    dia = fecha[i][5:7]
    año = fecha[i][9:13]
    mes = meses[mes]
    fecha[i] = año + '-' + mes + '-'+ dia
```

```python
list_all_data=[]
for i in range(len(fecha)):
    list_all_data.append([fecha[i],valores[i*6+0],valores[i*6+1],
    valores[i*6+2],valores[i*6+3],valores[i*6+4],valores[i*6+5]])
```

```python
df_mana=pd.DataFrame(list_all_data)
```
